```python
import re
import pandas as pd
from pathlib import Path

input_csv = Path("SUB789-11_17_MANIFEST.csv")
output_csv = Path("SUB789-11_17_MANIFEST_fixed.csv")

df = pd.read_csv(input_csv)

pattern = re.compile(
    r"Subject(?P<subject>\d+)Activity(?P<activity>\d+)Trial(?P<trial>\d+)Camera(?P<camera>\d+)/(?P<filename>[^/]+)$"
)

def fix_image_path(path):
    text = str(path).replace("\\", "/")
    match = pattern.search(text)
    if not match:
        return path

    subject = int(match.group("subject"))
    if subject < 11:
        return path

    return (
        f"Subject{subject}/"
        f"Activity{int(match.group('activity'))}/"
        f"Trial{int(match.group('trial'))}/"
        f"Camera{int(match.group('camera'))}/"
        f"{match.group('filename')}"
    )

df["image_path"] = df["image_path"].apply(fix_image_path)

df.to_csv(output_csv, index=False)

print(f"Saved: {output_csv}")
print(df["image_path"].head())
```
