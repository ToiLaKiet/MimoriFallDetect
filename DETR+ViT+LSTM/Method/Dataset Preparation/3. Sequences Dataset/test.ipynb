```python
import json
import pandas as pd
json_data = ''
with open('sequences_data.json', 'r') as f:
    json_data = json.load(f)
```

```python
val = []
nonval = []
for element in json_data:
    if ((pd.Series(element['label_array']).nunique()) > 1 and (element['Subject'] in ['Subject7', 'Subject8', 'Subject9','Subject11'])):
        val.append(element)
    elif (element['Subject'] in ['Subject7', 'Subject8', 'Subject9','Subject11']):
        nonval.append(element)
```

```python
print(len(nonval),len(val))
```
